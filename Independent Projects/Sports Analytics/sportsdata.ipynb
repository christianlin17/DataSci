```python
import pandas as pd
import numpy as np
import seaborn as sns 
from matplotlib import pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('rush_def.csv')

df.head()
```

```python
xs = df['2019']
ys = df['Rank']

plt.plot(xs, ys, 'o')
plt.xlabel('Last year defense stats')
plt.ylabel('2020 Rankings')
plt.show

res = linregress(xs, ys)
print(res)

fx = np.array([xs.min(), xs.max()])
fy = res.intercept + res.slope*fx
plt.plot(fx, fy, '-')

```

```python
# How strongly correlated is a Defense's Overall Rush Defense Performance with their Last 3 Week's 2020 Performance?
x2 = df['2020']
y2 = df['Last 3']

plt.plot(x2, y2, 'o')
plt.xlabel('2020 Defense')
plt.ylabel('2020 Last 3 Weeks')
plt.show

res2 = linregress(x2, y2)
print(res2)

fx = np.array([x2.min(), x2.max()])
fy = res2.intercept + res2.slope*fx
plt.plot(fx, fy, '-')
```
